```python
def ModelTrain():
    

    train_dir="F:\OPENCV\data\data1"


    train_datagen=ImageDataGenerator(rescale=1./255,
                                       shear_range=0.2,
                                     rotation_range=40,
                                     fill_mode='nearest',
                                       zoom_range=0.2,
                                       validation_split=0.2, # validation split 10%.
                                       horizontal_flip=True)
    
    img_width,img_height =125,125
    input_shape=(img_width,img_height,3)
    batch_size =20
# change
    train_generator =train_datagen.flow_from_directory(train_dir,
                                                   target_size=(img_width,img_height),
                                                   batch_size=batch_size)

    num_classes= len(train_generator.class_indices)

    print(train_generator.class_indices)
    
    vgg=VGG16(input_shape=input_shape,weights='imagenet',include_top=False)
    
    for l in vgg.layers:
        l.trainable=False
        
    x=Flatten()(vgg.output)
    
    
    prd=Dense(2,activation='softmax')(x)
    model=Model(inputs=vgg.input,outputs=prd)
    
    model.summary()
    
   
  
    
    validation_generator = train_datagen.flow_from_directory(
        train_dir, # same directory as training data
        target_size=(img_height, img_width),
        batch_size=batch_size) 

#     # Model building to get trained with parameters.
    opt=keras.optimizers.Adam(lr=0.001)
    model.compile(optimizer=opt,loss='categorical_crossentropy',metrics=['accuracy'])
    train=model.fit_generator(train_generator,
                              nb_epoch=5,
                              steps_per_epoch=train_generator.samples // batch_size,
                              validation_data=validation_generator,
                             nb_val_samples= validation_generator.samples// batch_size,verbose=1)
    
    
    return model
```
